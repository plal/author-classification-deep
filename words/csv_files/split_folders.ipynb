```python
import os

files = os.listdir(r'../png_files/')
len(files)
```

```python

all_data = []
for file in files:
    if len(file.split('-')[1]) < 4: print(file)
    label = file[:3] + file.split('-')[1][-1]
    all_data.append({'file':file, 'label':label})
```

```python
len(all_data)
```

```python
import pandas as pd

all_df = pd.DataFrame(all_data)
all_df
```

```python
all_df.to_csv('all_new.csv', index=False)
```
